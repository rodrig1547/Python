```python
train_df['Sex'].replace(['female', 'male'], [0,1], inplace= True)
test_df['Sex'].replace(['female', 'man'], [0,1], inplace= True)
train_df['Embarked'].replace(['Q','S','C'],[0,1,2], inplace = True)
test_df['Embarked'].replace(['Q','S','C'],[0,1,2], inplace = True)

##___________________________________
## Reemplazo los datos faltantes en la edad por la media de esta columna
##___________________________________
print (train_df['Age'].mean())
print (test_df['Age'].mean())

train_df['Age'].replace( np.nan ,train_df['Age'].mean())
test_df['Age'].replace( np.nan, test_df['Age'].mean())

##___________________________________
## Creo Varias grupos de acuerdo a bandas de las edades
##___________________________________

bins = [0,8,15,18,25,40,60,100]
names = ['1','2','3','4','5','6','7']
train_df['Age'] = pd.cut(train_df['Age'], bins, labels=names)
test_df['Age']= pd.cut(test_df['Age'], bins, labels=names)

##___________________________________
## Eliminar columnas innecesarias
##___________________________________

train_df =train_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
test_df =test_df.drop(['Name', 'Ticket'], axis = 1)
##___________________________________
## Eliminar datos faltantes
##___________________________________
train_df.dropna(axis = 0, how = 'any', inplace= True )
test_df.dropna(axis = 0, how = 'any', inplace= True )




```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Separación la columna con la información de los sobrevivientes
X = np.array(train_df.drop(['Survived'],axis =1))
y = np.array(train_df['Survived'])

X_train, X_test, y_train, y_test = train_test_split(X ,y , test_size= 0.2)
```

```python
## Regresión logistica
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
print ('Precisión Regresión Logística')
print (logreg.score(X_train, y_train))
```
